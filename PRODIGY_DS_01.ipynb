**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, Taking exapmle of a netflix IMDB  dataset.**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/IMDB-Movie-Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#beginning the plot
data=pd.DataFrame(df)

fig=plt.figure()
ax=fig.add_axes([0,0,3.5,3.5])
x=list(data.iloc[:0])
y=list(data.iloc[:1])
ax.bar(x,y,color='b')
plt.show
```
